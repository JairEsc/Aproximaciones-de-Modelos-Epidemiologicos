Dependencia de los tiempos desde última reacción para el tiempo del siguiente evento de un proceso de renovación de tiempos Weibull

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
ax = plt.figure(figsize=(10,14)).add_subplot(projection='3d')

S = np.arange(0, 2, 0.05)
T = np.arange(0, 2, 0.05)
S,T = np.meshgrid(S,T)
l=1
a=1.5
Z = l*a*(l*(S+T))**(a-1)*np.exp((l*T)**(a)-(l*(S+T))**(a))# Plot the 3D surface
ax.plot_surface(S,T, Z, edgecolor='black', lw=0.7, rstride=4, cstride=4,
                alpha=0.1)
ax.set_xlabel('s',weight='bold',labelpad=6,fontsize=12)
ax.set_ylabel(r'$\bf{\tilde{t}}$',labelpad=10,fontsize=12)
ax.set_zlim3d()
ax.set_zlabel(r'$\tilde{\psi}\left(s+\tilde{t}|\tilde{t}\right)$',labelpad=8,fontsize=12,weight = 'bold')
# Make data.
ax.set_box_aspect(aspect=None, zoom=0.8)
# Plot a basic wireframe.
plt.savefig('no_markoviano.png',dpi=300)
plt.show()
```

```python
ax = plt.figure(figsize=(10,14)).add_subplot(projection='3d')

S = np.arange(0, 2, 0.05)
T = np.arange(0, 2, 0.05)
S,T = np.meshgrid(S,T)
l=1
a=1.5
Z = l*np.exp(-l*S)
ax.plot_surface(S,T, Z, edgecolor='black', lw=0.7, rstride=4, cstride=4,
                alpha=0.1)
ax.set_xlabel('s',weight='bold',labelpad=6,fontsize=12)
ax.set_ylabel(r'$\bf{\tilde{t}}$',labelpad=10,fontsize=12)
ax.set_zlim3d()
ax.set_zlabel(r'$\tilde{\psi}\left(s+\tilde{t}|\tilde{t}\right)$',labelpad=8,fontsize=12,weight = 'bold')
# Make data.
ax.set_box_aspect(aspect=None, zoom=0.8)
# Plot a basic wireframe.
plt.savefig('markoviano.png',dpi=300)
plt.show()
```
